### class `TabularDataset`

```python
from sklearn.datasets import load_boston
from src.core import data
```

#### training/evaluation dataset

```python
 X, y = load_boston(return_X_y=True)
cates_feats = np.random.choice((0, 1, 2), size=X.shape[0])

print('X : ', X.shape)
print('cates_feats : ', cates_feats.shape)
print(cates_feats)

cate_X = np.hstack((X, cates_feats.reshape(-1, 1)))
print(cate_X.shape)
```

```python
dataset = data.TabularDataset(feats=cate_X, targets=y)
```

```python
n_limit = 10

for i in range(len(dataset)):
    _x, _y = dataset[i]
    print(_x)
    print(_y)
```

#### inference dataset

```python
infer_dataset = data.TabularDataset(feats=cate_X, targets=None)
```

```python
n_limit = 10

for i in range(len(infer_dataset)):
    _x = infer_dataset[i]
    print(_x)
```

### func `create_data_loader`

```python
data_loader = data.create_data_loader(cate_X, y, batch_size=32, is_shuffle=True, num_workers=1, pin_memory=False)
```

```python
for loaded_data in data_loader:
    batch_X, batch_y = loaded_data
    print(batch_X.size())
    print(batch_y.size())
```
