```python
import torch
import numpy as np
from sklearn.datasets import load_boston
from tabnet.core._models import EmbeddingEncoder
from tabnet.utils.utils import CatePreprocessor
```

```python
X, _ = load_boston(return_X_y=True)
cates_feats = np.random.choice((0.2, 0.221, 0.222222, 0.12345), size=X.shape[0])

print('X : ', X.shape)
print('cates_feats : ', cates_feats.shape)
print(cates_feats)

cate_X = np.hstack((X, cates_feats.reshape(-1, 1)))
print(cate_X.shape)

input_dims = cate_X.shape[1]
cate_indices = [13]
# cate_dims = [3]
embed_dims = 1

preprocessor = CatePreprocessor()
processed_X, cate_dims = preprocessor.fit_infer(cate_X, cate_indices=[13])


embedding_encoder = EmbeddingEncoder(
    input_dims, cate_indices, cate_dims, embed_dims
)

embeds = embedding_encoder(torch.from_numpy(processed_X))

print(embeds.size())
```
