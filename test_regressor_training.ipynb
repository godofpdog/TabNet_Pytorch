```python
import numpy as np 
import pandas as pd 

from src.utils import mkdir
from src.logger import init_logger
from src.builtin.estimators import TabNetRegressor
from sklearn.datasets import load_boston
```

```python
logger_dir = 'logs'
logger_name = 'TestRegression'
level = 'INFO'

logger = init_logger(logger_dir=logger_dir, logger_name=logger_name, level=level)
```

```python
X, y = load_boston(return_X_y=True)

print(X.shape)
print(y.shape)
```

```python
tabnet = TabNetRegressor(
    input_dims=13, output_dims=[1], logger=logger, is_cuda=True,
    reprs_dims=4, atten_dims=4, num_steps=3, num_indep=2, num_shared=1
)
```

```python
tabnet.build(path=None)
```

```python
from torch.optim import Adam

optimizer_params = {'lr': 2e-2}
```

```python
tabnet.fit(X, y.reshape(-1, 1), batch_size=506, metrics=['mse'], max_epochs=200, optimizer=Adam, optimizer_params=optimizer_params)
```

```python
m_explain, masks = tabnet.explain(X)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(4, 1, figsize=(15,15))

for i in range(3):
    axs[i].imshow(masks[i].cpu().numpy()[:10])
    axs[i].set_xlabel('features')
    axs[i].set_ylabel('samples')
    axs[i].set_title(f"mask {i}")

axs[3].imshow(m_explain.cpu().numpy()[:10])
axs[3].set_xlabel('features')
axs[3].set_ylabel('samples')
axs[3].set_title('importance')
```
